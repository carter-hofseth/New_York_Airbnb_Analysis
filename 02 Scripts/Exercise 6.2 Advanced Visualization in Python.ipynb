### Correlation heatmap

```python
# Create dataframe with only neccessary columns 
sub = df_bnb_clean[['neighbourhood_group','neighbourhood','room_type','price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']]
```

```python
sub
```

```python
# Create correlation chart of our sub dataframe
sub.corr()
```

```python
# Create empty subplot for heatmap
f,ax = plt.subplots(figsize=(12,12))
```

```python
# Create a correlation heatmap with seaborn
corr_heat = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

```python
corr_heat.figure
```

The coefficients within the matrix represent how strong their linear relationship is with each other. Positive means that it has a direct relationship, when one variable goes up the other goes up to and vice versa. The negative means it has an inverse relationship, when one variable goes up, the other variable goes down and vice versa. 

In terms of this data set, there are not a lot of strong relationships between the columns. Except reviews and reviews per month, which is to be expected since they are similar variables. There is a small realtionship between availability and host listings that I would like to investigate further.

```python
# Create sub dataframe with just numeric columns
sub_mat = df_bnb_clean[['price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']]
```

## Categorial plots

```python
# Create dataframe with no prices above $2000
df_price_check = df_bnb_clean
```

```python
df_price_check.loc[df_price_check['price'] > 500, 'price'] = np.nan
```

```python
df_price_check['price'].max()
```

```python
# Create histogram of distribution of price
sns.distplot(df_price_check['price'], bins = 100)
```

```python
sub.loc[sub['price'] < 50, 'Price category'] = 'Low price'
```

```python
sub.loc[(sub['price'] >= 50) & (sub['price'] < 500), 'Price category'] = 'Normal price'
```

```python
sub.loc[sub['price'] >= 500, 'Price category'] = 'High price'
```

```python
sub['Price category'].value_counts(dropna = False)
```

```python
# Create categorial plot using price categories created above

sns.set(style='ticks')
c = sns.catplot(x = 'minimum_nights',y='price', hue = 'Price category', data = sub)
```

Well, from here we can see that a there are a lot of high priced Airbnbs that are skewing our data. So we might have to delete these outliers later on in our analysis but for now I will keep them until our analysis progresses.
