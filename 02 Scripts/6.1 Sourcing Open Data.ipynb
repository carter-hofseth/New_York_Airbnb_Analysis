### Mixed data types

```python
# Find mixed data types in data set
for col in airbnb_ny_df.columns.tolist():
    werid = (airbnb_ny_df[[col]].applymap(type) != airbnb_ny_df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (airbnb_ny_df[werid]) > 0:
        print(col)
```

```python
# Get the data types of the mixed data types
airbnb_ny_df['name'].dtypes
airbnb_ny_df['host_name'].dtypes
airbnb_ny_df['last_review'].dtypes
```

## Data Cleaning


```python
# Fill in missing values with N/A
airbnb_ny_df['name'].fillna('N/A', inplace = True)
airbnb_ny_df['host_name'].fillna('N/A', inplace = True)
```

```python
# Check to see if code went through
airbnb_ny_df['name'].value_counts(dropna = False)
```

```python
airbnb_ny_df['reviews_per_month'].value_counts(dropna = False)
```

```python
airbnb_ny_df['last_review'].value_counts(dropna = False)
```

```python
airbnb_ny_df['host_name'].value_counts(sort=True, dropna = False)
```

### Mixed data

```python
# change data types of mixed data columns (name to string, host_name to string, last_review to datetime)
airbnb_ny_df['name'] = airbnb_ny_df['name'].astype('str')
```

```python
airbnb_ny_df['host_name'] = airbnb_ny_df['host_name'].astype('str')
```

```python
airbnb_ny_df['last_review'] = pd.to_datetime(airbnb_ny_df['last_review'])
```

```python
# Check the data types of columns
airbnb_ny_df.info()
```

```python
# Check for mixed data again
for col in airbnb_ny_df.columns.tolist():
    werid = (airbnb_ny_df[[col]].applymap(type) != airbnb_ny_df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (airbnb_ny_df[werid]) > 0:
        print(col)
```

### Change column names

```python
# Change "name" to "airbnb_name"
airbnb_ny_df.rename(columns = {'name':'airbnb_name'},inplace = True)
```
